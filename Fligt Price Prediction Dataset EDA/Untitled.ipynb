```python
# Feature Engineering 
```

### Splitting the Date_of_Journey col with new cols - Date, Month and Year

```python
final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
```

```python
final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
final_df['Month'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
final_df['Year'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Date'].astype(int)
final_df['Year'] = final_df['Date'].astype(int)
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head()
```

### Splitting the Arrival_Time by only extracting the Arrival Time not the month and date

```python
final_df['Arrival_Time'].apply(lambda x : x.split(' '))
```

```python
final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

### Splitting the Arrival_Time with new cols - Arrival_Hour and Arrival_Min

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df.head(2)
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_Min'] = final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df.head(2)
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Min'] = final_df['Arrival_Hour'].astype(int)
```

```python
final_df.drop('Arrival_Time',  axis=1, inplace=True)
```

```python
final_df.head(2)
```

### Splitting the Dep_Time with new cols - Dep_Hour and Dep_Hour

```python
final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
final_df['Dep_Min'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Min'] = final_df['Dep_Min'].astype(int)
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head(2)
```

```python
final_df.info()
```

### Total Stops

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

### Converting Duration

```python
final_df['Duration_Hour'] = final_df['Duration'].apply(lambda x : x.split(' ')[0].split('h')[0])
```

```python
final_df[final_df['Duration_Hour'] == '5m']
```

```python
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

```python
final_df['Duration_Hour'] = final_df['Duration_Hour'].astype(int)
```

```python
final_df.drop('Duration', axis=1, inplace=True)
```

```python
final_df.head(2)
```

### Label Incoding for Airline

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
